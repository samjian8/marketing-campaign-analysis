Exploring the Data

```python
df['PercentIncomeSpent'] = ((df['MntTotal'] / df['Income']) / 2) * 100 # Dividing by two since the amounts spent are for the previous TWO years
df.head(20)
```

```python
df['MntTotal'].hist(bins=20)
plt.show()
```

```python
df.columns
```

```python
sum_values = {
    'Wines': df['MntWines'].sum(),
    'Fruits': df['MntFruits'].sum(),
    'Meat': df['MntMeatProducts'].sum(),
    'Fish': df['MntFishProducts'].sum(),
    'Sweets': df['MntSweetProducts'].sum(),
    'Gold': df['MntGoldProds'].sum()
}
```

```python
ax = pd.Series(sum_values).plot(kind='bar', title='Two Years\' Spending By Product Type')
ax.set_xlabel('Product Type')
ax.set_ylabel('Total Amount Spent in Last Two Years (Dollars)')
```

```python
pd.Series(sum_values).sum()
```

```python
((df['MntWines'].sum() + df['MntMeatProducts'].sum()) / pd.Series(sum_values).sum()) * 100
```
