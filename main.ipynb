It appears that there is an outlier in the data. We will attempt to isolate this outlier by the interquartile method

```python
Q1 = df['MntMeatProducts'].quantile(0.25)
Q3 = df['MntMeatProducts'].quantile(0.75)

IQR = Q3 - Q1

is_outlier = (df['MntMeatProducts'] < (Q1 - 1.5 * IQR)) | (df['MntMeatProducts'] > (Q3 + 1.5 * IQR))

outlier_df = df[is_outlier]
print(is_outlier.sum())
```

It seems that we have other outliers. However, these may still be relevant to our data. We will isolate this particular outlier in the top-right region of the graph by another method.


```python
sns.boxplot(x=df['MntMeatProducts'])
```

```python
Q1 = df['MntMeatProducts'].quantile(0.25)
Q3 = df['MntMeatProducts'].quantile(0.75)

IQR = Q3 - Q1

is_outlier_extended = (df['MntMeatProducts'] < (Q1 - 5 * IQR)) | (df['MntMeatProducts'] > (Q3 + 5 * IQR))

outlier_df = df[is_outlier_extended]
print(outlier_df)
```

```python
sns.scatterplot(x='Response',
                y='Income',
                data=df)
plt.show()
```

```python
summary_stats = df.groupby('Response')['Income'].describe()
print(summary_stats)
```

```python
perecent_diff_target = ((summary_stats.loc[1, 'mean'] - summary_stats.loc[0, 'mean']) \
                        / summary_stats.mean().mean()) * 100
print(perecent_diff_target)
```

It seems that, on average, those who responded to the target marketing campaign were those of higher income than those who did not respond. In fact, those who responded had a 22.4% higher income on average. But how does this compare to customers who responded (or didn't respond) to any campaign(s) in general?

```python
df['AcceptedCmpAny'] = df['AcceptedCmpOverall'].apply(lambda x: 0 if x == 0 else 1)
df.head(15)
```

```python
print(df.groupby('AcceptedCmpAny')['Income'].describe())
```

```python
summary_stats1 = df.groupby('AcceptedCmpAny')['Income'].describe()
print(summary_stats1)
```

```python
perecent_diff_other = ((summary_stats1.loc[1, 'mean'] - summary_stats1.loc[0, 'mean']) \
                        / summary_stats1.mean().mean()) * 100
print(perecent_diff_other)
```

It seems that customers who responded to at least 1 campaign had an income level 37.1% higher on average compared to non-responders. Perhaps future marketing campaigns should be targeted towards higher-income grouops. We will continue to find other groups that responded highly to marketing campaigns.
